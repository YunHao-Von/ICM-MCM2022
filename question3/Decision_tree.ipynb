```python
import pandas as pd
data = pd.read_csv('useful_data/all_data.csv')[['Country Name','Country Code','GNI',
                                                'Percent_Of_Educate','CPIA','Percent_Energy','Female_Labor_Percent',
                                                'Poor_Percent','Per_Water','Gini','Malnutrition',
                                                'CPI score 2021','port','Value']].iloc[:260,]
data = data[~(data['GNI'].isnull())] 
data
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
features = data[['Percent_Of_Educate','CPIA','Percent_Energy','Female_Labor_Percent',
                'Poor_Percent','Per_Water','Gini','Malnutrition',
                'CPI score 2021','port','Value']]
label = data[['GNI']]
model = DecisionTreeRegressor(max_depth=10)
features = features.fillna(0)
model.fit(features, label)
```

```python
import graphviz
from sklearn.tree import export_graphviz
feature_names = ['Percent_Of_Educate','CPIA','Percent_Energy','Female_Labor_Percent',
                'Poor_Percent','Per_Water','Gini','Malnutrition',
                'CPI score 2021','port','Value']
dot_data=export_graphviz(model,feature_names=feature_names,out_file=None)
dot=graphviz.Source(dot_data)
dot.view()
```
